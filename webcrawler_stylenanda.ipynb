```python
from selenium import webdriver
url = 'http://stylenanda.com/#none'

driver = webdriver.Chrome('chromedriver')
driver.get(url)
```

```python
search_button = driver.find_element_by_xpath('//*[@id="sb-icon-search_"]/img')
search_button.click()
```

```python
context = driver.find_element_by_xpath('//*[@id="keyword"]')
context.clear()
context.send_keys('bag')
```

```python
search_button = driver.find_element_by_xpath('//*[@id="searchBarForm"]/div/fieldset/input[2]')
search_button.click()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
info_all = soup.find_all('li', class_='item xans-record-')
item_list = []
price_list = []
for info in info_all:
    tmp = info.find('div', class_='name')
    item = tmp.find_all('span')
    item_list.append(item[1].text)
    price  = info.find('p', class_='price')
    price_list.append(price.text.split()[0])
```

```python
nextpage = driver.find_element_by_xpath('//*[@id="paging"]/ul/li[2]/a')
nextpage.click()
```
